```python


from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df2 = df1 
```

```python

le.fit(df1['Partner'])
Partner = le.transform(df1['Partner'])
df2 = df2.drop(['Partner'], axis=1)
df2.insert(1,'Partner',Partner)

le.fit(df1['PaperlessBilling'])
PaperlessBilling = le.transform(df1['PaperlessBilling'])
df2 = df2.drop(['PaperlessBilling'], axis=1)
df2.insert(1,'PaperlessBilling',PaperlessBilling)

le.fit(df1['TechSupport'])
TechSupport = le.transform(df1['TechSupport'])
df2 = df2.drop(['TechSupport'], axis=1)
df2.insert(1,'TechSupport',TechSupport)

le.fit(df1['StreamingTV'])
StreamingTV = le.transform(df1['StreamingTV'])
df2 = df2.drop(['StreamingTV'], axis=1)
df2.insert(1,'StreamingTV',StreamingTV)

le.fit(df1['StreamingMovies'])
StreamingMovies = le.transform(df1['StreamingMovies'])
df2 = df2.drop(['StreamingMovies'], axis=1)
df2.insert(1,'StreamingMovies',StreamingMovies)

le.fit(df1['Contract'])
Contract = le.transform(df1['Contract'])
df2 = df2.drop(['Contract'], axis=1)
df2.insert(1,'Contract',Contract)

le.fit(df1['DeviceProtection'])
DeviceProtection = le.transform(df1['DeviceProtection'])
df2 = df2.drop(['DeviceProtection'], axis=1)
df2.insert(1,'DeviceProtection',DeviceProtection)

le.fit(df1['OnlineBackup'])
OnlineBackup = le.transform(df1['OnlineBackup'])
df2 = df2.drop(['OnlineBackup'], axis=1)
df2.insert(1,'OnlineBackup',OnlineBackup)
le.fit(df1['MultipleLines'])
MultipleLines = le.transform(df1['MultipleLines'])
df2 = df2.drop(['MultipleLines'], axis=1)
df2.insert(1,'MultipleLines',MultipleLines)

le.fit(df1['OnlineSecurity'])
OnlineSecurity = le.transform(df1['OnlineSecurity'])
df2 = df2.drop(['OnlineSecurity'], axis=1)
df2.insert(1,'OnlineSecurity',OnlineSecurity)

le.fit(df1['Dependents'])
Dependents = le.transform(df1['Dependents'])
df2 = df2.drop(['Dependents'], axis=1)
df2.insert(1,'Dependents',Dependents)

le.fit(df1['InternetService'])

InternetService = le.transform(df1['InternetService'])
df2 = df2.drop(['InternetService'], axis=1)
df2.insert(1,'InternetService',InternetService)


le.fit(df1['PhoneService'])
PhoneService = le.transform(df1['PhoneService'])
df2 = df2.drop(['PhoneService'], axis=1)
df2.insert(1,'PhoneService',PhoneService)

le.fit(df1['MultipleLines'])
MultipleLines = le.transform(df1['MultipleLines'])
df2 = df2.drop(['MultipleLines'], axis=1)
df2.insert(1,'MultipleLines',MultipleLines)

le.fit(dataset['PaymentMethod'])
PaymentMethod = le.transform(dataset['PaymentMethod'])
df2 = df2.drop(['PaymentMethod'], axis=1)

df2.insert(1,'PaymentMethod',PaymentMethod)
```
